# Numerical Categorical
1. bargraph
2. boxplot
3. violinplot
4. stripplot
5. swarmplot

```python
sns.barplot(data=df,x='sex',y='total_bill',hue='day')
plt.show()
```

```python
plt.bar(df['sex'],df['total_bill'])
plt.show()
```

```python
g1=df.groupby('sex')
```

```python
g1.groups
```

```python
df.groupby('sex')['total_bill'].sum()
```

```python
plt.figure(figsize=(10,6))
df.groupby('sex')['total_bill'].sum().plot(kind='bar')
plt.show()
```

### Boxplot

```python
sns.boxplot(data=df,x='day',y='total_bill',palette='Pastel2_r');
plt.title('day vs total bill')
plt.show()
```

### Vilionplot

```python
sns.violinplot(data=df,x='time',y='tip',hue='sex')
plt.grid()
plt.show()
```

```python
sns.violinplot(data=df,x='smoker',y='total_bill',hue='sex', split=True)
plt.show()
```

```python
sns.violinplot(data=df,x='day',y='total_bill',hue='sex',split=True)
plt.show()
```

# Stripplot

```python
sns.stripplot(data=df,y='total_bill',x='day')
plt.axhline(20,color='k')
plt.show()
```

### Swarmplot

```python
sns.swarmplot(data=df,x='day',y='total_bill')
plt.axhline(20,color='k')
plt.show()
```

### Categorical Categorical

```python
df.columns
```

```python
pd.crosstab(df['sex'],df['smoker'])
```

```python
pd.crosstab(df['sex'],df['smoker']).plot(kind='bar')
plt.grid()
plt.show()
```

```python
sns.countplot(data=df,x='sex',hue='smoker')
plt.show()
```

### Stacked Bar Graph

```python
pd.crosstab(df['day'],df['smoker'])
```

```python
pd.crosstab(df['day'],df['smoker']).plot(kind='bar',stacked=True)
plt.show()
```
