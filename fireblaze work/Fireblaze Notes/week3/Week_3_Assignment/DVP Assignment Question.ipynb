<span style='color:Red'> **About The Dataset(StudentsPerformance)** </span>

This data set consists of the marks secured by the students in various subjects.

* **lunch:** standard, free/reduced

* **math score:** subject marks

### <span style='color:Blue'> **Question - 25. Is there any relationship between math and reading score? If yes? show the answer and explain it.** </span>

```python
sns.jointplot(data=dtf, x="math score", y="reading score")
sns.set_style('darkgrid') 
plt.show()
```

In the above example we create Bivarient Scatterplot with Marginal Univarient Histograms. The Scatterplot shows us the relationship between different Math score and Reading score, 
which is math score around 40-85 and reading score 40-85 show good results,
and Histograms with the same relationship improve our assesment better.
