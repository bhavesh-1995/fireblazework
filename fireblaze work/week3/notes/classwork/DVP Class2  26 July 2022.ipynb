# Scatter plot
- Scatter plot is a graph in which the values of two variables are plotted along two axes. It is a most basic type of plot that helps you visualize the relationship between two variables

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
plt.scatter(df['total_bill'],df['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Relation between total_bill vs tip')
plt.show()
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip')
plt.show()
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='day')
plt.show()
```

# LM Plot (Linear model Plot)
- lmplot is one of the most basic plots. It shows a line on a 2 dimensional plane. 
- The lmplot in seaborn fit regression models with intercept.

```python
sns.lmplot(data=df,x='total_bill',y='tip')
plt.show()
```

```python
sns.lmplot(data=df,x='total_bill',y='tip',hue='time')
plt.show()
```

# Regplot
- This method is used to plot data and a linear regression model fit

```python
sns.regplot(data=df,x='total_bill',y='tip')
plt.show()
```

# Pairplot
- To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function

```python
df.head()
```

```python
sns.pairplot(df)
plt.show()
```

# Jointplot
- It is used to draw a plot of two variables with bivariate and univariate graphs. It basically combines two different plots.

```python
sns.jointplot(data=df,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',kind='reg',color='r')
plt.show()
```

```python
sns.jointplot(data=df,x='total_bill',y='tip')
plt.show()
```

# Heatmap
- Heatmap is defined as a graphical representation of data using colors to visualize the value of the matrix.
- In this, to represent more common values or higher activities brighter colors basically reddish colors are used and to represent less common or activity values, darker colors are preferred. 
- Heatmap is also defined by the name of the shading matrix.
- Heatmaps in Seaborn can be plotted by using the seaborn.heatmap() function.

```python
# Range -1 to 1
df.corr()
```

```python
sns.heatmap(df.corr(),annot=True,cmap='Accent')
plt.show()
```

- Line plot is a type of chart that displays information as a series of data points connected by straight line segments. A line plot is often the first plot of choice to visualize any time series data.
